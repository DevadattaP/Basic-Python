**Perform various operations on data using Pandas Library**

```python
import pandas
```

```python
data = [10,20,30,40]
series1 = pandas.Series(data)
print(series1)
```

```python
print(series1[0:2])
```

```python
data = [[1,2,3,4],[5,6]]
df = pandas.DataFrame(data)
print(df)
```

```python
data = {'a':[1,2,3,4],
        'b':[5,6,7,8]}
df = pandas.DataFrame(data)
print(df)
```

```python
data = {'Name':['Devadatta','Suyog','Vansh','Sumit'],
        'Roll No.':[50,24,14,52]}
df = pandas.DataFrame(data)
print(df)
```

```python
data = {'a':[1,2,3,4],
        'b':[5,6,7,8]}
df1 = pandas.DataFrame(data)
data = {'Name':['Devadatta','Suyog','Vansh','Sumit'],
        'Roll No.':[50,24,14,52]}

df2 = pandas.DataFrame(data)
df3 = pandas.concat([df1,df2])
print(df3)
```

```python
df4 = pandas.concat([df1,df2], axis=1, join='inner')
print(df4)
```

```python
data = {'a':[1,2,3,4],
        'Roll No.':[50,24,14,52]}
df1 = pandas.DataFrame(data)
data = {'Name':['Devadatta','Suyog','Vansh','Sumit'],
        'Roll No.':[50,24,14,52]}

df2 = pandas.DataFrame(data)
df3 = pandas.merge(df1,df2,how='inner')
print(df3)
```
