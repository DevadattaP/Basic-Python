**Creating class and objects**

```python
class Student:
    name = 'abc'
    roll = 0

print('student s1:')
s1 = Student()
print(s1.name,s1.roll)
print('student s2:')
s2 = Student()
s2.name='Devadatta'
s2.roll=50
print(s2.name,s2.roll)
```

**Using default constructor**

```python
class Student:
    def __init__(self):
        self.name = 'Devadatta'
        self.roll = 50
    def show(self):
        print('Hi, my name is',self.name)
        print('My roll number is',self.roll)

s1 = Student()
s2 = Student()
print('Student s1 :')
s1.show()
print('Student s2 :')
s2.show()
```

**Using parameterized constructor**

```python
class Student:
    def __init__(self,x,y,z):
        self.name = x
        self.roll = y
        self.marks= z
    def show(self):
        print('Hi, my name is',self.name)
        print('My roll number is',self.roll)
        print('I got',self.marks,'marks')

s1 = Student('Devadatta',50,20)
s2 = Student('Sumit',52,19)
print('Student s1 :')
s1.show()
print('Student s2 :')
s2.show()
```

**Q. Define a python class that has two methods- 'get_string' and 'print_uppercase'.**

```python
class myString:
    def get_string(self,st):
        self.stg=st
    def print_uppercase(self):
        print('Your string in uppercase is',str(self.stg).upper())
s1 = input('Enter any string : ')
str1 = myString()
str1.get_string(s1)
str1.print_uppercase()
```
