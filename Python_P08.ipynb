**1. Duck typing in python**

```python
class specialstring:
    def __len__(self):
        return 50

str1 = specialstring()
print(len(str1))
```

**2. Operator overloading**

Overloading '+' operator

```python
# using + on integers to add them
print(5 + 10)
# using + on strings to concatenate them
print('Devadatta '+'Pokharanakar')
# using + on lists to join them
print([1,2,3]+[4,5,6])
```

Overloading '*' operator

```python
# using * on integers to add them
print(5 * 10)
# using * on strings to repeat them
print('Devadatta ' * 3)
# using * for power operation 
print(3**3)
```

**3. Method overloading**

Finding sum of multiple numbers through method

```python
class myclass:
    result = 0
    def sum(self,a=None,b=None,*args):
        if a!=None and b!=None:
            result = a + b
            for i in args:
                result += i
            print('Addition is',result)
        else:
            print('Need at least 2 parameters')

obj =myclass()
obj.sum(3,4)
obj.sum(3,4,5)
obj.sum(3,4,5,6,7)
obj.sum(1)
```

**4. Method overriding**

```python
class Bird:
    def intro(self):
        print('There are many birds.')
    def fly(self):
        print('Some birds can fly but some cannot.')

class Sparrow(Bird):
    def fly(self):
        print('Sparrow can fly.')

class Ostrich(Bird):
    def fly(self):
        print('Ostrich cannot fly.')
    
obj_bird=Bird()
obj_spr=Sparrow()
obj_ost=Ostrich()

obj_bird.intro()
obj_bird.fly()

obj_spr.intro()
obj_spr.fly()

obj_ost.intro()
obj_ost.fly()
```
