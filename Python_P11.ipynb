**Linked list in Python**

```python
ll=[]
ll.append("India") 
ll.append("Japan") 
ll.append("America") 
print('the existing list =',ll)
# display menu 
choice =0 
while choice<5:
    print('\n** Linked list operation **')
    print('1 Add element')
    print('2 Remove element')
    print('3 Replace element')
    print('4 search element') 
    print('5 Exit element') 
    choice=int(input('\nEnter your choice : '))
    if choice==1: 
        element=input('\nEnter element : ') 
        pos=int(input('At what position : ')) 
        ll.insert(pos,element)
    elif choice==2: 
        try: 
            element=input('\nEnter element : ') 
            ll.remove(element) 
        except ValueError: 
            print('\nElement not found')
        
    elif choice==3: 
        element=input('\nEnter element : ') 
        pos=int(input('At what position : ')) 
        ll.pop(pos) 
        ll.insert(pos,element)
    elif choice==4: 
        element=input('\nEnter element : ') 
        try: 
            pos=ll.index(element) 
            print('Element found at position',pos) 
        except ValueError: 
            print('Element not found')
    
    else: 
        break 
print('List =',ll)
```
