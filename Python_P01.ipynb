**Q 9. Write a code to perform bitwise operations (Complement, AND and OR, Left shift, Right Shift)**

```python
a = 5
b = 4
print('Bitwise complement: ~',a ,'=',~a)
print('Bitwise and:',a ,'&',b,'=',a&b)
print('Bitwise or:',a ,'|',b,'=',a|b)
print('Bitwise xor:',a ,'^',b,'=',a^b)
print('Bitwise left shift:',a ,'<<',b,'=',a<<b)
print('Bitwise right shift:',a ,'>>',b,'=',a>>b)
```
