# **Q 7. Write a code to find largest number in list.**

```python
lis=[6,4,7,2,9,1,3]
max=lis[0]
for i in lis:
  if i>max:
    max=i
print('Maximum number in list is',max)
```

# **Q 8. Write a code to remove duplicate number in list.**

```python
dup = [1,2,3,1,2,3,4,5,2,3,5]
ndup=[]
for i in dup:
  if i not in ndup:
    ndup.append(i)
print(ndup)
```

# **Q 10. Write a code for writing a phone number in words using dictionary.**

```python
phone=input('Enter your mobile number : ')
num_map={'1':'one',
         '2':'two',
         '3':'three',
         '4':'four',
         '5':'five',
         '6':'six',
         '7':'seven',
         '8':'eight',
         '9':'nine',
         '0':'zero'}
for i in phone:
  print(num_map.get(i),end='\t')
```
