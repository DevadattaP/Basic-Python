**Q7. Write a code for factorial of a number with and without recursive function.**

Without recursion

```python
def fact(num):
    f=1
    for i in range(1,num+1):
        f*=i
    return f
n=int(input('ENter a number: '))
print('factorial of',n,'is',fact(n))
```

with recursion

```python
def factorial(num):
    if num==0 or num==1:
        return 1
    else:
        return num*factorial(num-1)
n=int(input('ENter a number: '))
print('factorial of',n,'is',factorial(n))
```

**Q10. Use function to count number of capital letters and small letters from given string.**

```python
def capital_small(str):
    small,capital=0,0
    for i in str:
        if(i.isupper()):
            capital+=1
        if(i.islower()):
            small+=1
    return small,capital
low,up=capital_small('DevaDatta')
print('Small letters =',low)
print('Capital letters =',up)
```
