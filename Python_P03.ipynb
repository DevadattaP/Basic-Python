**1. If else statement**

```python
temp = int(input('Enter temperature: '))
if temp<=20:
    print("Its cold day.")
else:
    print("Its not cold today.")
```

**2. If elif else**

```python
name = input('Enter your name: ')
if len(name)<3:
    print("Name cannot be less than 3 characters.")
elif len(name)>20:
    print("Name cannot exceed 20 characters.")
else:
    print("Name has correct length.")
```

**3. Nested if**

```python
marks = int(input('Enter your marks: '))
if marks>40:
    if marks<50:
        print("D grade")
    elif marks<60:
        print("C grade")
    elif marks<75:
        print("B grade")
    elif marks<85:
        print("A grade")
    else:
        print("O grade")
else:
    print("F grade")
```

**4. While Loop**

```python
i=1
while i<=5:
    print(i)
    i+=1
print('The end')
```

```python
i=1
while i<=5:
    print('*'*i)
    i+=1
```

```python
i=1
while i<=5:
    print(' '*(5-i),'*'*i)
    i+=1
```

```python
i=1
while i<=5:
    print(' '*(5-i),'* '*i)
    i+=1
```

**6. Break, continue and pass**

```python
for letter in 'Devadatta':
    if letter=='a':
        break
    else:
        print(letter)
```

```python
for letter in 'Devadatta':
    if letter=='a':
        continue
    else:
        print(letter)
```

```python
for letter in 'Devadatta':
    if letter=='a':
        pass
        print(' PASS')
    else:
        print(letter)
```
